**Student Marks Analyzer**

**Problem Statement**

Develop a Python program using NumPy to analyze student marks. The program should accept marks of multiple students and store them in a NumPy array. It should calculate the total marks, average marks, highest marks, and lowest marks. Based on the average marks, the program should assign grades to students using student-defined grading rules. Conditional statements must be used to implement the grade logic and display the final results clearly.

**Breif Explanation of Code**

This program analyzes student marks using Python and NumPy. It first takes the number of students as input and stores their marks in a list, which is then converted into a NumPy array for easier processing. A loop is used to calculate the total marks, while NumPy functions are used to find the average, highest, and lowest marks. Conditional statements are applied to assign grades (A, B, C, or Fail) to each student based on their individual marks. Finally, the program displays the grades of all students along with the overall marks analysis.

```python
import numpy as np

students = int(input("Enter number of students: "))  # Accept number of students.

list = []  # List to store marks.

for i in range(students):  # Taking marks input.
    marks = float(input(f"Enter marks of student {i+1}: "))
    list.append(marks)

marks_array = np.array(list)  # Convert list to NumPy array.
total_marks = 0
for i in range(students):  # Addition of student marks.
  total_marks += marks_array[i]
average_marks = np.mean(marks_array)  # Average
highest_marks = np.max(marks_array)  # Highest
lowest_marks = np.min(marks_array)  # Lowest
for i in range(students):

  if marks_array[i] > 85: # Grade logic.
      grade = "A"
  elif marks_array[i] >= 70:
      grade = "B"
  elif marks_array[i] >= 50:
      grade = "C"
  else:
      grade = "Fail"
  print(f"Student {i+1} Grade:", grade)

print("\n----- STUDENT MARKS ANALYSIS -----")  # Display results.
print("Students Total Marks:", total_marks)
print("Students Average Marks:", average_marks)
print("Students Highest Marks:", highest_marks)
print("Students Lowest Marks:", lowest_marks)

```
